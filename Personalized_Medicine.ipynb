```python
import pandas as pd
import numpy as np
import re
from pm_utils import *
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
train_text = pd.read_csv('data/training_text', sep="\|\|", engine='python', header=None, skiprows=1, names=["ID","Text"])
train_variants = pd.read_csv('data/training_variants')

train_df = pd.merge(train_variants, train_text, on='ID')
m = train_df.shape[0]
del train_text, train_variants
```

```python
word_to_index, index_to_word, word_to_vec_map = read_glove_vecs('data/glove.6B.50d.txt')
```

```python
# Cleaning up the text
for i in range(m):
    txt = train_df.iloc[i, train_df.columns.get_loc('Text')].lower()
    txt = re.sub(r'\.', ' . ', txt)
    txt = re.sub(r'\/|\-', " ", txt)
    train_df.iloc[i, train_df.columns.get_loc('Text')] = txt
```

```python
# Generating training data

word = word_tokenize(train_df.iloc[0,train_df.columns.get_loc('Text')].lower())
```
